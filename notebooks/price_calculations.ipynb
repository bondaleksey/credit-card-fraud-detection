# Оценка стоимости эксплуатации 

## Data Proc
Тарифы эксплуатации ресурсов 
Data Proc в [yandex cloud](https://cloud.yandex.ru/docs/data-proc/pricing)

Для кластера Data Proc с двумя подкластерами со следующими характеристиками:
- мастер-подкластер: класс хоста s3-c2-m8, размер хранилища 40 ГБ;
- data-подкластер: класс хоста s3-c4-m16, 3 хоста, размер хранилища
128 ГБ;
стоимость использования составляет 36р в час.

![alt cluster](img/dataproc_pricing.png "Data Proc cost")

То есть месячное использование такого кластера составит:
$36*24*31 = 26 784$, безучета исходящий трафика из Yandex Cloud в интернет.


## S3
Здесь приведены [тарифы на s3](https://cloud.yandex.ru/docs/storage/pricing) в yandex.

## Итого
Итого оценочная стоимость месячного использования кластера + S3 (- без учета исходящего трафика вне инфраструктуры yandex cloud) составляет:
$27 127р$


# Варианты оптимизации
## S3
Использование Холодного хранилища, при "незначительной" (100 000 GET-операции и POST-операции) нагрузке на обращения к S3 приведет к снижению цены эксплуатации 
до 288р, то есть составит 84% от цены хранения в стандартном хранилище (343р). 

Однако, при увеличении частоты обращений более 200 000 GET-операции и POST-операции, экономически выгоднее использовать стандартное хранилище.


## DataProc
Для оптимизации надо сделать следующие шаги:
 - преобразовать данные из txt в parquet;
 - рассмотреть возможность использования только compute узлов, вместо data узлов, то есть возможность отказаться от hdfs и перейти только на работу с данными только из s3;
 - исходя из вычислительных потребностей, найти "удовлетворяющие" вычислительным задачам характеристики worker узлов кластера.
